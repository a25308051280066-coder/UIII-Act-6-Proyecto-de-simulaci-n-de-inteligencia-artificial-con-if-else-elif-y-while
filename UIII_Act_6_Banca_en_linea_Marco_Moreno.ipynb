### Banca en linea


```python
# ======================================================
# 🏦 CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Marco Moreno
# 🏦 Proyecto: IA de Banca en Línea
# ======================================================

print("========= SISTEMA IA DE BANCA EN LÍNEA 🏦 =========")

while True:

    nombre = input("\nIngresa nombre del cliente: ")

    saldo = int(input("Ingresa saldo disponible: "))

    retiro = int(input("Ingresa monto del retiro: "))

    intentos = int(input("Ingresa cantidad de intentos fallidos: "))

    verificacion = input("¿Cuenta verificada? si/no: ")

    # ======================================================
    # 🧠 Inteligencia Artificial
    # ======================================================

    if saldo >= retiro:

        print("Retiro aprobado correctamente")

    else:

        print("Saldo insuficiente")

    if retiro >= 10000:

        print("Movimiento bancario alto detectado")

    elif retiro >= 5000:

        print("Movimiento bancario medio detectado")

    else:

        print("Movimiento bancario pequeño detectado")

    if verificacion == "si" and intentos == 0:

        print("Cuenta segura")

    elif intentos >= 1 and intentos <= 3:

        print("Actividad sospechosa detectada")

    else:

        print("Cuenta bloqueada temporalmente")

    # ======================================================
    # 📊 Resultado final
    # ======================================================

    print("\n========= RESULTADOS =========")

    print("Cliente:", nombre)

    print("Saldo:", saldo)

    print("Retiro:", retiro)

    print("Intentos fallidos:", intentos)

    print("Cuenta verificada:", verificacion)

    continuar = input("\n¿Deseas analizar otra cuenta? si/no: ")

    if continuar == "no":

        print("\nPrograma finalizado 🏦")

        break
```
